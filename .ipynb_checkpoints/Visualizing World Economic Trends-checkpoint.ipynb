```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import csv
import os

country_profiles_file = "country_profile_variables.csv"

```

```python
# loading csv file and putting into dataframe

country_data = pd.read_csv(country_profiles_file)
country_data.head()
```

```python
# Cleaning up the data!

environ_df = country_data.loc[:,["country", 
                                 "Region",
                                 "GDP: Gross domestic product (million current US$)", 
                                 "GDP per capita (current US$)",'CO2 emission estimates (million tons/tons per capita)']]

environ_df = environ_df.set_index("GDP: Gross domestic product (million current US$)")
environ_df = environ_df.drop('-99', axis = 0)

environ_df = environ_df.reset_index()
environ_df = environ_df.set_index("CO2 emission estimates (million tons/tons per capita)")
environ_df = environ_df.drop('-99', axis = 0)

environ_df = environ_df.rename(columns={"country":"Country"})
environ_df = environ_df.reset_index()
environ_df = environ_df.set_index("Country")

environ_df.rename(columns={"CO2 emission estimates (million tons/tons per capita)":"CO2 emission estimates (million tons)"}, inplace= True)


cleaned_environ_df = environ_df[["Region", "GDP: Gross domestic product (million current US$)", 
                         "GDP per capita (current US$)",
                         "CO2 emission estimates (million tons)"]]

# Grouping regions together to create cleaner graphs
cleaned_environ_df["Region"] = cleaned_environ_df["Region"].replace(["MiddleAfrica","EasternAfrica", "WesternAfrica", "NorthernAfrica", "SouthernAfrica"], "Africa")
cleaned_environ_df["Region"] = cleaned_environ_df["Region"].replace(["CentralAsia", "EasternAsia", "South-easternAsia", "WesternAsia", "NorthernAsia", "SouthernAsia"], "Asia")
cleaned_environ_df["Region"] = cleaned_environ_df["Region"].replace(["EasternEurope", "WesternEurope", "NorthernEurope", "SouthernEurope"], "Europe")
cleaned_environ_df["Region"] = cleaned_environ_df["Region"].replace(["SouthAmerica","CentralAmerica"], "SouthAmerica")


# cleaned_environ_df["Region"].value_counts()
cleaned_environ_df.head()
# print(cleaned_regions)
```

```python
# Grouping by region and running aggregate functions

grouped_env = cleaned_environ_df.groupby("Region")
grouped_gdp = grouped_env["GDP per capita (current US$)"].sum()
grouped_co2 = round(grouped_env["CO2 emission estimates (million tons)"].sum(), 2)

gdp_co2_df = pd.DataFrame(grouped_gdp)
gdp_co2_df["CO2 emission estimates (million tons)"] = grouped_co2

# Changing names to increase readability
gdp_co2_df.rename({"CentralAmerica": "Central America", 
               "CentralAsia": "Central Asia",
               "EasternAfrica": "Eastern Africa",
               "EasternAsia": "Eastern Asia",
               "EasternEurope":"Eastern Europe",
               "MiddleAfrica":"Middle Africa",
               "NorthernAfrica":"Northern Africa",
               "NorthernAmerica":"North America",
               "South-easternAsia": "Southeast Asia",
               "SouthernAsia": "Southern Asia",
               "SouthAmerica":"South America",
               "SouthernAfrica":"Southern Africa",
               "SouthernEurope": "Southern Europe",
               "WesternAfrica": "Western Africa",
               "NorthernEurope": "Northern Europe",
               "WesternAsia": "Western Asia",
               "WesternEurope": "Western Europe"
              }, inplace = True)

gdp_co2_df = gdp_co2_df.reset_index()
gdp_co2_df.set_index("Region", inplace = True)

# removing outliers

gdp_co2_df.drop("Micronesia", axis=0, inplace = True)
gdp_co2_df.drop("Polynesia", axis=0, inplace = True)
gdp_co2_df.drop("Melanesia", axis=0, inplace = True)
gdp_co2_df.drop("Caribbean", axis=0, inplace = True)
gdp_co2_df.drop("Oceania", axis=0, inplace = True)

gdp_co2_df = gdp_co2_df.sort_values("CO2 emission estimates (million tons)")
gdp_co2_df = gdp_co2_df.sort_values("GDP per capita (current US$)")
gdp_co2_df = gdp_co2_df.reset_index()




# co2_energy_df = pd.DataFrame(grouped_co2)
# print(grouped_gdp)
# print(grouped_species)
gdp_co2_df.head()
# co2_energy_df.head()
```

```python
gdp_co2_bar = gdp_co2_df.plot(x= gdp_co2_df["Region"],
                              kind="bar", 
                              figsize = (10,5), 
                              title = "Total GDP Per Capita (USD) vs. Total CO2 Emission Estimate (Million Tons)",
                              rot = 0)
plt.rcParams['font.size'] = 13.0

plt.savefig("gdp_co2_bar.png")
```

```python
gdp_list = list(gdp_co2_df["GDP per capita (current US$)"])
co2_list = list(gdp_co2_df["CO2 emission estimates (million tons)"])
zippedRatio_list = [round(a/b, 2) for a,b in zip(gdp_list, co2_list)]

region_list = list(gdp_co2_df["Region"])
ratio_df = pd.DataFrame(region_list)
ratio_df["GDP:CO2 ratio"] = zippedRatio_list

colors = ["lightblue", "lightcoral", "lightgreen", "yellow", "orange"]
explode = (0,0,0,0,0.05)

plt.pie(zippedRatio_list, explode = explode, labels = region_list, colors= colors, shadow = True, startangle = 110, autopct='%1.1f%%')
plt.title("Identifying the Most Carbon Efficient Economies")
plt.axis("equal")

plt.savefig("ratio.png")
# ratio_df.head()
```
